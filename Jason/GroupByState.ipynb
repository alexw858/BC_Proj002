```python
state_group = fema_df.groupby(["state"])
state_group.sum()
```

```python
project_amount = state_group['projectAmount'].sum()
fed_funded = state_group['federalShareObligated'].sum()

state_summary_table = pd.DataFrame({"ProjectAmount":project_amount,
                                   "FederalFunded":fed_funded})

```

```python
state_summary_table["%Funded"] = state_summary_table["FederalFunded"] / state_summary_table["ProjectAmount"]
```

```python
state_summary_table.reset_index(inplace=True)
state_summary_table
```

```python
state_summary_table.to_csv("state_funded.csv", index=False)
```

```python
weather_state = fema_df.groupby(['state','incidentType']).count()
state_incident_count = pd.DataFrame({"ProjectAmount":weather_state['dcc']})
```

```python
state_incident_count.reset_index(inplace=True)
```

```python
state_incident_count.to_csv("incidenttype_count.csv", index=False)
```

```python
incident_type = fema_df.groupby(["incidentType"])
incident_type.sum().head()
```

```python
incident_amount = incident_type['projectAmount'].sum()
fed_funded_incident = incident_type['federalShareObligated'].sum()

incident_summary_table = pd.DataFrame({"ProjectAmount":incident_amount,
                                   "FederalFunded":fed_funded_incident})

incident_summary_table["%Funded"] = incident_summary_table["FederalFunded"] / incident_summary_table["ProjectAmount"]

incident_summary_table.reset_index(inplace=True)
```

```python
incident_summary_table.to_csv("incident_funded.csv", index=False)
```
